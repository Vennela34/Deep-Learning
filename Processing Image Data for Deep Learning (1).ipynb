**Resizing the image using Pillow library**

```python
from PIL import Image
```

```python
img = Image.open('/content/tiger.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('tiger_image_resized.jpg')
```

```python
img_res = mpimg.imread('/content/tiger_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Converting RGB images to Grayscale image using OpenCV**

```python
import cv2
```

```python
img = cv2.imread('/content/tiger.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
cv2.imwrite('tiger_grayscale_image.jpg', grayscale_image)
```
